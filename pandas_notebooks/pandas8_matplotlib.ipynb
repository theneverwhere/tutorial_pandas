# Построение графиков и визуализация

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# для более функциональной работы с окнами графиков прописать:|
# %matplotlib notebook 

# установка стиля графиков
plt.style.use('seaborn-whitegrid') # стиль графиков 
```

```python
x = np.array([-3,-2,-1,0,1,2,3])
y = np.array([9,4,1,0,1,4,9])
```

## Рисунки и подграфики

```python
fig1, ax1 = plt.subplots()  # задаёт полотно (рисунок) и сам график
ax1.plot(x,y);  # обращаемся к графику и вносим данные
```

```python
# можно отдельно создать рисунок с графиками

fig_1 = plt.figure() # есть множество параметров!

# figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None)
# num - номер рисунка, figsize - (float, float) ширина и высота в дюймах, dpi - разрешение, facecolor -цвет фона,
# edgecolor - цвет краёв 
```

```python
# и отдельно добавлять подграфики
ax_1 = fig_1.add_subplot(2,3,6)  # рисунок будет расчерчен в сетке 2х3 и вставляем подграфик в место 6 
ax_2 = fig_1.add_subplot(2,2,1)  # тот же рисунок абстрактно делится сеткой 2х2 и вставляется подграфик в место 1 и тд
ax_3 = fig_1.add_subplot(2,2,3)  
ax_4 = fig_1.add_subplot(2,2,2)
fig_1
```

Эти объекты осей графика обладают различными методами, которые создают графики различных типов. Рекомендуется использовать именно методы осей, а не верхнеуровневые функции построения типа plt.plot.

```python
ax_1.plot(x,y,color='red')    # у графиков есть множество параметров, такие как color, linestyle, alpha
ax_2.scatter(x,y,c='r')
ax_3.plot(np.random.standard_normal(50), linestyle='--')
ax_4.plot(np.random.standard_normal(100).cumsum(), alpha=0.15)
fig_1
```

```python
fig, axes = plt.subplots(2,2)  # но удобнее не по отдельности, а через subplots
```

```python
axes[1,0].plot(np.random.standard_normal(15))  # обращение к подграфику во второй строчке первой по счету [1,0]
fig
```

#### 3адание свободного места вокруг подграфиков
Величину промежутка легко изменить с помощью метода subplots_adjust объекта Figur

```python
# По умолчанию matplotlib оставляет пустое место вокруг каждого подграфика и между подграфиками
fig_1.subplots_adjust(wspace=0)
fig_1.subplots_adjust(hspace=0)
fig_1
```

## Цвета, маркеры и стили линий

```python
fig, ax = plt.subplots()
# можно обозначать параметры кратко
ax.plot(x,y,color='green',linestyle='-')
ax.plot(x,y*2, c='r',ls='--')
ax.plot(x,y*0.5,color='0.1',linestyle=':', linewidth=5)
ax.plot(x*0.2,y, color='#AAFF33', lw=3)
```

```python
# установка маркеров точек
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(20).cumsum(), c='black',marker='o')


# быстрый доступ к параметрам через короткую запись

ax.plot(x,y, 'r--*')
```

По умолчанию на линейных графиках соседние точки соединяются отрезками прямой, т.  е. производится линейная интерполяция. Параметр drawstyle
позволяет изменить этот режим:

```python
fig, ax = plt.subplots()
data = np.random.standard_normal(20).cumsum()
ax.plot(data ,drawstyle='steps-post',ls='--')
ax.plot(data, c='black',ls=':')
```

### Задание названия графика, названий осей, рисок и их меток

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(100).cumsum())  # случайное блуждание

ax.set_xticks([0,25,50,75,100])    # установить риски
ax.set_yticks(np.arange(-5,10,1));
ax.set_xticklabels(['one','two','three','four','five'],rotation=30, fontsize=8)   # смена названия меток рисок
```

Подписи и метки

```python
fig, ax = plt.subplots()
ax.plot([1,3,-4,2,11,0,-3,5])   # или установить параметр label=
ax.plot(x,y,'r:o')

ax.set_title('ПРИВЕТ Я ЗАГОЛОВОК')
ax.set_xlabel('НАЗВАНИЕ ОСИ Х')
ax.set_ylabel('НАЗВАНИЕ ОСИ Y')
ax.legend(['f','h'])


ax.spines['right'].set_visible(False)   # справа нет границы )
ax.grid(color = 'blue', linewidth = 0.5, linestyle = '--') # как будет расчерчена сетка (чтобы убрать сетку ax.grid())
```

## Аннотации и рисование в подграфике
В matplotlib имеются объекты, соответствующие многим стандартным фигурам, они называются патчами (patches). Часть из них, например Rectangle, Circle и Polygon, находится в модуле matplotlib.pyplot, а весь набор – в модуле matplotlib.patches.

```python
fig, ax = plt.subplots()
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
 color="green", alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

Помимо стандартных типов графиков, разрешается наносить на график свои 
аннотации, которые могут включать текст, стрелки и другие фигуры. Для добавления аннотаций и  текста предназначены функции text, arrow и  annotate.

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.text(1,3,'туть')  # текст в заданных коорлинатах
ax.arrow(2,2,0,4,)  # типа стрелка
ax.annotate('Привет мир, я аннотация для этой замечательной параболы!',(-3,8))
```
